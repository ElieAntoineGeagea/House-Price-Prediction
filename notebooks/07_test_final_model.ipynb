```python
import pandas as pd
import numpy as np
import joblib

# Load final saved model
model = joblib.load("../models/final_ridge_model.joblib")

# Load processed training data only to get the correct columns
X_train_processed = pd.read_csv("../data/processed/X_train_processed.csv")

# Create one fake house example using median values as a safe base
example_house = X_train_processed.median().to_frame().T

# Modify important features to create a realistic example
example_house["OverallQual"] = 7
example_house["GrLivArea"] = np.log1p(1800)
example_house["GarageCars"] = 2
example_house["GarageArea"] = 500
example_house["TotalBsmtSF"] = np.log1p(900)
example_house["1stFlrSF"] = np.log1p(900)
example_house["FullBath"] = 2
example_house["YearBuilt"] = 2005
example_house["YearRemodAdd"] = 2005
example_house["TotalSF"] = example_house["TotalBsmtSF"] + example_house["1stFlrSF"] + example_house["2ndFlrSF"]
example_house["TotalBath"] = 2
example_house["HouseAge"] = 2010 - 2005
example_house["RemodAge"] = 2010 - 2005
example_house["HasGarage"] = 1
example_house["HasBasement"] = 1
example_house["HasPool"] = 0

# Predict log price
log_prediction = model.predict(example_house)

# Convert back to normal dollar price
price_prediction = np.expm1(log_prediction)

print("Predicted house price:", price_prediction[0])
```
